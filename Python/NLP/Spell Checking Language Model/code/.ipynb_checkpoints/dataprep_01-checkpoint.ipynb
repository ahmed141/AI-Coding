```python
import codecs
import numpy as np
import glob
```

# Reading text file

```python
file_name = "../data/jang.txt"
```

```python
f_words1 = codecs.open(file_name, "r", "utf-8")
train_data = f_words1.read()
```

```python
train_tokens = train_data.split('\n')
```

```python
train_set = [0]*(len(train_tokens) - (len(train_tokens)//5))
test_set = [0]*(len(train_tokens)//5)
```

```python
print(len(train_set))
print(len(test_set))
print(len(test_set) + len(train_set))
```

```python
len(train_tokens)
```

```python
i_test = 0
i_train = 0
for i in range(0, len(train_tokens)):
    print("iteration#\t", i)
    if (i%5 == 0):
        test_set[i_test] = train_tokens[i]
        i_test += 1
    else:
        train_set[i_train] = train_tokens[i]
        i_train += 1
```

```python
print(train_set[0])
print(test_set[0])
```

```python
train = ''

for i in range(0, len(train_set)):
    train += train_set[i]
    train += '\n'
```

```python
held_out = ''

for i in range(0, len(test_set)):
    held_out += test_set[i]
    held_out += '\n'
```

```python
file = codecs.open('../data/Q01_jang_train.txt', 'w', 'utf-8')
file.write(train)
file.close()
```

```python
file = codecs.open('../data/Q01_jang_heldout.txt', 'w', 'utf-8')
file.write(held_out)
file.close()
```
