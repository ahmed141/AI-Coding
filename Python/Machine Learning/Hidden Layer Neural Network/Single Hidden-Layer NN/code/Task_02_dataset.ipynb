```python
import numpy as np
import random
import matplotlib.pyplot as plt
#To make it compatible with jupyter notebook
%matplotlib inline
# Fixing random state for reproducibility
np.random.seed(1)
```

```python
# < ------------- GENERATING DATA FOR TASK-2 ---------------------- >
dims = 2
num_samples = 400
#Creating two circles
from sklearn.datasets import make_circles

X, Y = make_circles(n_samples = num_samples, factor = .4, noise = 0.05)

#Lets visualize the dataset
reds = Y == 0
blues = Y == 1
Y.shape = (Y.shape[0], 1)
'''
plt.scatter(X[reds, 0], X[reds, 1], c = "red",s = 20, edgecolor = 'k')
plt.scatter(X[blues, 0], X[blues, 1], c = "blue", s = 20, edgecolor = 'k')

plt.show()
#plt.savefig("./images/Task02/Dataset.png")
'''
```

```python
'''
print(X.shape)
print(Y.shape)
print(X[reds].shape)
print(X[blues].shape)

print(Y[reds].shape)
print(Y[blues].shape)
'''
```

```python
sample_num_foreach = X[reds].shape[0]
```

```python
# < ------------- Spliting 60%, 20%, 20% ---------------------- >
#Data spliting
#60% training, 20% validation, 20% test
train_page = int(0.6*sample_num_foreach)   #first 60%
val_page = train_page + int(0.2*sample_num_foreach)  #next 20%
test_page = val_page + int(0.2*sample_num_foreach)   #next 20%
'''
print(train_page)
print(val_page)
print(test_page)
'''
```

```python
# train data, first 120
train_X = X[reds][0:train_page, :]
train_X = np.concatenate( (train_X, (X[blues][0:train_page, :])), axis = 0)

train_Y = Y[reds][0:train_page, :]
train_Y = np.concatenate( (train_Y, (Y[blues][0:train_page, :])), axis = 0)
```

```python
'''
print(train_X.shape)
print(train_Y.shape)
'''
```

```python
# validation data, second 40
val_X = X[reds][train_page:val_page, :]
val_X = np.concatenate( (val_X, (X[blues][train_page:val_page, :])), axis = 0)

val_Y = Y[reds][train_page:val_page, :]
val_Y = np.concatenate( (val_Y, (Y[blues][train_page:val_page, :])), axis = 0)
```

```python
'''
print(val_X.shape)
print(val_Y.shape)
'''
```

```python
# test data, next 40
test_X = X[reds][val_page:test_page, :]
test_X = np.concatenate( (test_X, (X[blues][val_page:test_page, :])), axis = 0)

test_Y = Y[reds][train_page:val_page, :]
test_Y = np.concatenate( (test_Y, (Y[blues][val_page:test_page, :])), axis = 0)
```

```python
'''
print(test_X.shape)
print(test_Y.shape)
'''
```

```python
plt.figure(figsize=(20,10))
plt.scatter(train_X[:, 0], train_X[:, 1], marker = '^', c = "red", s = 30, label = "Train")
plt.scatter(val_X[:, 0], val_X[:, 1], marker = 'o', c = "blue", s = 30, label = "Validation")
plt.scatter(test_X[:, 0], test_X[:, 1], c = "yellow", s = 30,  label = "Test")
plt.legend(loc='best')

plt.show()
#plt.savefig("./images/Task01/DatasetSplit.png")
```
