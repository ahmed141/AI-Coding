```python
import codecs
import numpy as np
import glob
import pickle
import numpy as np
```

```python
Train_model = pickle.load(open("../data/train_model3.0v2.pkl", 'rb'))
```

```python
def read_file(doc_name):
    f = open(doc_name, 'r')
    f_str = f.read()
    f_words = f_str.replace('<', ' < ').replace('.', ' . ').replace('"', ' " ').replace(':', ' : ').replace('#', ' # ').replace(',', ' , ').replace('!', ' ! ').replace('%', ' % ').split()
    
    return f_words
```

```python
def test_bayesian(model, doc_name):
    f_words = read_file(doc_name)
    
    ham_prob = 0
    spam_prob = 0
    
    for i in f_words:
        if i in model['spam_vocab']:
            if (spam_prob != 0):
                spam_prob *= model['spam_prob'][model['spam_vocab'].index(i)]
            else:
                spam_prob = model['spam_prob'][model['spam_vocab'].index(i)]
            if (ham_prob != 0):
                ham_prob *= model['ham_prob'][model['ham_vocab'].index(i)]
            else:
                ham_prob = model['ham_prob'][model['ham_vocab'].index(i)]
        
    spam_prob *= model['spam_prior']
    ham_prob *= model['ham_prior']
    
    class_name = None
    if (spam_prob > ham_prob):
        class_name = "spam"
    else:
        class_name = "ham"
    
    out = {
        'ham_prob': ham_prob,
        'spam_prob': spam_prob,
        'class_name': class_name
    }
    return out
```

```python
path_spam_docs = glob.glob("../data/test/spam/*.txt")
path_ham_docs = glob.glob("../data/test/ham/*.txt")

tp = 0
fp = 0
tn = 0
fn = 0

ct = 0
for i in path_ham_docs:
    test = test_bayesian(Train_model, i)
    print('++', ct, test['class_name'])
    ct += 1
    
    if( test['class_name'] == 'ham' ):
        tp += 1
    else:
        fp += 1
ct = 0
for i in path_spam_docs:
    test = test_bayesian(Train_model, i)
    print('--', ct, test['class_name'])
    ct += 1
    
    if( test['class_name'] == 'spam' ):
        tn += 1
    else:
        fn += 1

```

```python
print("Truly classified Spam:\t\t", tn)
print("Truly classified Ham:\t\t", tp)
print("Falsely classified Spam:\t", fn)
print("Falsely classified Ham:\t\t", fp)
```

```python
accuracy = (tp+tn)/(tp+tn+fp+fn)
recall = tp/(tp+fn)
precision = tp/(tp+fp)
F1 = 2*((recall*precision)/(recall+precision))

print("Accuracy:\t", accuracy)
print("Precision:\t", precision)
print("Recall:\t\t", recall)
print("F1 Score:\t", F1)
```

```python
Results = {
    'tp': tp,
    'tn': tn,
    'fp': fp,
    'fn': fn,
}
```

```python
pickle.dump(Results, open( "../data/Result.pkl", "wb"))
```
