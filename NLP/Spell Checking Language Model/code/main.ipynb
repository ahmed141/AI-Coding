# Using All Functions

```python
error_words = []
for i in range(0, len(error_tokens)):
    if(error_tokens[i] != nonerror_tokens[i]):
        error_words.append(i)
total_error_words = len(error_words)
```

```python
print("Error words count:\t", total_error_words)
```

```python
found = 0
notfound = 0
f = codecs.open("../results/report.txt", "w+", "utf-8")
for i in range(0, total_error_words):
    print("error token#\t\t", i)
    er_words = error_tokens[error_words[i]-1]+' '+error_tokens[error_words[i]]
    actual_word = nonerror_tokens[error_words[i]]
    print("error word:\t\t", error_tokens[error_words[i]])
    print("actal word:\t\t", actual_word)
    
    cans, probs = get_rank10_candidates(er_words, Letters, words_list, unigram_prob, bigram_prob, lambda1=0.7, lambda2=0.3)
    
    filestr = ''
    filestr += 'input error word: \t\t\t\t '+error_tokens[error_words[i]]+'\n'
    filestr += 'top-10 candidates:\n'
    for i in range(0, len(cans)):
        filestr += str(cans[i]) + '\t\t\t\t\t\t\t' + str(probs[i]) + '\n'
    filestr += 'Remarks:\t\t\t\t\t\t'
    
    if (actual_word in cans):
        print("actual found in candidates\n")
        filestr += 'Found in Candidates list\n'
        found +=1
    else:
        print("actual NOT found in candidates\n")
        filestr += 'NOT Found in Candidates list\n'
        notfound +=1

    filestr += 'Actual Word:\t\t\t\t\t'+actual_word
    filestr += '-----------------------------------------------------------------------------------------------\n'
    f.write(filestr)
f.close()
```
