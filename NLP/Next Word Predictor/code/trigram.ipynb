## Evaluation

```python
file_name = "../data/allsentences_test.urdu"

f_words1 = codecs.open(file_name, "r", "utf-8")
test_data = f_words1.read()
```

```python
test_data = test_data.split()
```

```python
test_data = test_data[0:int(len(test_data)/4)]
```

```python
len(test_data)
```

```python
keys_normal = 0
for i in range(0, len(test_data)):
    keys_normal += len(test_data[i])
```

```python
keys_normal
```

```python
count_ = 1
str_word = ''

for i in range(0, len(test_data)):
    print(i, "--->", test_data[i])
    for j in range(0, len(test_data[i])):
        str_word += test_data[i][j]
        
        temps = suggest_words_trigram(trigram_prob, word_ = str_word, n = 5)
        
        if (test_data[i] in temps):
            break
        else:
            count_ += 1
```

```python
keys_pred = count_
```

```python
print("KS = ", (((keys_normal - keys_pred)/keys_normal)*100), "%")
```
