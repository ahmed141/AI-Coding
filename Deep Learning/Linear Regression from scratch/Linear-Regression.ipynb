## Linear Regression Implementation

```python
def LinearRegressionparms( x, y ):
    '''
   Parameters:
   x - vector with x points
   y - vector with y points
   Returns:
   m - slope of line
   c - y-intercept
   '''

    A = np.ones((x.shape[0], 2))
    A[:, 0] = x.T
    m, c = np.linalg.lstsq(A, y)[0]
    
    return m, c
```

### For DataSet_01

```python
m_01, c_01 = LinearRegressionparms(x_train_01, y_train_01)
print("m = " + str(m_01))
print("c = " + str(c_01))
xi_01 = np.arange(x_train_01.min()-1, x_train_01.max()+1)
yi_01 = (m_01)*xi_01 + c_01
```

```python
plt.figure(figsize=(20,10))

plt.plot(x_train_01, y_train_01, 'b.', label = "Train Dataset_01")
plt.plot(x_test_01, y_test_01, 'y.', label = "Test Dataset_01")
plt.plot(xi_01, yi_01, 'r-')
plt.legend()
print("The data points for Dataset_01 are shown in the plot")
```

## Mean Squared Error

```python
def MeanSqE( m, c, x, y):
    '''
   Parameters:
   m - slope of line
   c - y-intercept of line
   x - vector with x points
   y - ground truth y points
   Returns:
   mean_sq_error - mean squared error
   '''
    y_pred = (m)*x + c
    error = y_pred - y
    sq_error = np.square(error)
    mean_sq_error = np.divide(np.sum(sq_error), x.shape[0])
    
    return mean_sq_error
```

### For DataSet_01 

```python
print(MeanSqE(m_01, c_01, x_test_01, y_test_01))
```
